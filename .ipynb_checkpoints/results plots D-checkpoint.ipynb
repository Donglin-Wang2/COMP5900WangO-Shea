```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot(loss, title):
    plt.plot(loss[0], label='kl')
    plt.plot(loss[1], label='l')
    plt.plot(loss[2], label='r')       
    plt.legend()
    plt.savefig(f'./results/plots/{title}.png')
    plt.title(f'{title} loss')
    plt.show()
```

```python
# adapted

kl = np.load('./results/vae_adapted/inception_kl_loss.npy')
l = np.load('./results/vae_adapted/inception_loss.npy')
r = np.load('./results/vae_adapted/inception_reconstruction_loss.npy')

plot([kl, l, r], 'Adapted Inception')

kl = np.load('./results/vae_adapted/mobilenet_kl_loss.npy')
l = np.load('./results/vae_adapted/mobilenet_loss.npy')
r = np.load('./results/vae_adapted/mobilenet_reconstruction_loss.npy')

plot([kl, l, r], 'Adapted Mobilenet')

kl = np.load('./results/vae_adapted/resnet_kl_loss.npy')
l = np.load('./results/vae_adapted/resnet_loss.npy')
r = np.load('./results/vae_adapted/resnet_reconstruction_loss.npy')

plot([kl, l, r], 'Adapted Resnet')

kl = np.load('./results/vae_adapted/vgg_kl_loss.npy')
l = np.load('./results/vae_adapted/vgg_loss.npy')
r = np.load('./results/vae_adapted/vgg_reconstruction_loss.npy')

plot([kl, l, r], 'Adapted VGG')
```
