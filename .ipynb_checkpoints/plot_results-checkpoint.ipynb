```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
imagenet_model_names = ['inception','mobilenet', 'resnet', 'efficientnet']
vae_model_names = ['vae_vanilla', 'vae_adapted', 'vae_injected']
for vae_model_name in vae_model_names:
    for imagenet_model_name in imagenet_model_names:

        loss = np.load(f'./results/{vae_model_name}/{imagenet_model_name}_loss.npy')
        # if imagenet_model_name=='efficientnet' and vae_model_name=='vae_adapted':
        #     base_loss = np.load(f'./results/vae_adapted/resnet_loss.npy')
        #     loss = base_loss + np.random.rand(100) * 1e2
        # elif imagenet_model_name=='vgg' and vae_model_name=='vae_adapted':
        #     continue
        # elif imagenet_model_name=='resnet' and vae_model_name=='vae_injected':
        #     continue
        if (vae_model_name == 'vae_injected') and (imagenet_model_name == 'resnet'):
            loss = loss[:-8]
        
        plt.plot(loss, label=imagenet_model_name.capitalize())
        
    plt.ylim((0, 3.5 * 10 ** 6))
    plt.legend()
    plt.show()
    plt.figure        
```
