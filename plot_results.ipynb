```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
imagenet_model_names = ['inception','mobilenet', 'resnet', 'efficientnet']
vae_model_names = ['vae_vanilla', 'vae_adapted', 'vae_injected']
for vae_model_name in vae_model_names:
    for imagenet_model_name in imagenet_model_names:
        loss = np.load(f'./results/{vae_model_name}/{imagenet_model_name}_loss.npy')
        loss /= 64
        plt.plot(loss, label=imagenet_model_name.capitalize())
        
    plt.ylim((0, (3.5 * 10 ** 6) / 64))
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.show()
    plt.figure        
```
